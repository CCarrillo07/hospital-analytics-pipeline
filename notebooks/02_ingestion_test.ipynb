## 2. Load patients CSV into the raw table

This cell reads the patients CSV file, adds the `source_file` metadata column, and loads the data into the PostgreSQL table `raw.patients`.

The purpose of this step is to test the ingestion process manually before moving the logic into the automated ingestion script.

````python
df = pd.read_csv(r"C:\Users\crist\Documents\hospital_analytics_project\data\raw\patients\patients00.csv")
df["source_file"] = "patients_manual.csv"

df.to_sql(
    "patients",
    engine,
    schema="raw",
    if_exists="append",
    index=False
)
````
